# Juntanto os arquivos - Prouni 2013 até 2019

```python
import pandas as pd
from glob import glob
```

## Lendo arquivos

```python
arquivos = sorted(glob('../Prouni/N_Datasets/*.csv'))
```

```python
arquivos
```

### Lendo arquivos para um dataframe

```python
Todos_Prouni = pd.concat((pd.read_csv(cont, sep = ';', encoding = 'utf-8') for cont in arquivos))
```

```python
Todos_Prouni
```

```python
Todos_Prouni.to_csv('../Prouni/Prouni_2013a2019.csv', index = False)
```
